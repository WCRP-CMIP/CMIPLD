```python
import cmipld 
import pandas as pd
import tqdm
import glob

from cmipld.utils import organisation
import re,json

from cmipld.utils.validate_json import JSONValidator
val = JSONValidator('')

df= pd.read_csv("/Users/daniel.ellis/Downloads/Organisations-Grid view.csv")
# df.columns


```

```python

missing = []
ror = []

for index, row in tqdm.tqdm(df.iterrows(), total=len(df)):
    acronym = row['Institution -acronym (multiple)'].split(' - ')[0]
    if ' ' in acronym: acronym = None
    
    name = row['Full name']
    
    country = row['Country']
    
    if not acronym:
        missing.append(dict(name=name,country=country,acronym=acronym))
        continue

    try:
        rt = organisation.search_ror(name,country.replace('USA','US'))

        if isinstance(rt, list) and len(rt):
            if isinstance(rt[0], dict):
                ror_data = rt[0]
                formatted = organisation.format_institution(ror_data, acronym=acronym, mytype='institution')
                path = f'./ror/{formatted["@id"]}.json'
                json.dump(formatted, open(path,'w'), indent=4)
            else:
                print(rt)
        else:
            print(f'Could not find ROR for {name} ({acronym}) in {country}: {rt}')
            missing.append(dict(name=name,acronym=acronym,country=country,ror=rt))
            
            
    except TypeError as e:    
        print(f'Error for {name} ({acronym}) in {country}: {e}')
        missing.append(dict(name=name,acronym=acronym,country=country,ror=str(e)))
    # validate all files afterwards
    
    
    

```
